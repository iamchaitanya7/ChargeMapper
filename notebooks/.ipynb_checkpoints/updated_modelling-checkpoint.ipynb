```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, balanced_accuracy_score
from tqdm.notebook import tqdm
from sklearn.utils import all_estimators
import glob
```

```python
data = pd.read_csv("C:/Users/ASUS/OneDrive/DataScience/FINAL-PROJECT/data-science-Optimal-EV-station-placement-main/main_raw/data/processed/all_city_data_with_pop.csv")
```

```python
def data_splitter(data, train_cities=None, test_cities=None, test_size=0.2, random_state=42):
    if train_cities is not None:
        train = data[data['city'].isin(train_cities)]
        test = data[data['city'].isin(test_cities)]
        X_train = train.drop(['city', 'geometry', 'EV_stations'], axis=1)
        y_train = train['EV_stations'].astype(int).apply(lambda x: 1 if x > 0 else 0)
        X_test = test.drop(['city', 'geometry', 'EV_stations'], axis=1)
        y_test = test['EV_stations'].astype(int).apply(lambda x: 1 if x > 0 else 0)
    else:
        X = data.drop(['city', 'geometry', "EV_stations"], axis=1)
        y = data['EV_stations'].apply(lambda x: 1 if x > 0 else 0)
        X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test
```

```python
# Split the data
X_train, X_test, y_train, y_test = data_splitter(data)
```

```python

# Logistic regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print("Logistic Regression Test Accuracy: ", logreg.score(X_test, y_test))
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Get all classification model classes
classifiers = all_estimators(type_filter='classifier')

# Initialize result table
results = []
models = {}

# Run models and collect results
for name, ClassifierClass in tqdm(classifiers):
    try:
        model = ClassifierClass()
        model.fit(X_train, y_train)
        models[name] = model
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')
        auc = roc_auc_score(y_test, y_pred)
        balanced_accuracy = balanced_accuracy_score(y_test, y_pred)

        # Append results
        results.append([name, accuracy, precision, recall, f1, auc, balanced_accuracy])
    except Exception as e:
        print(f"Error occurred for {name}: {str(e)}")

# Create a DataFrame from results
results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-score", "AUC", "Balanced Accuracy"])
results_df = results_df.sort_values(by=['F1-score', 'AUC'], ascending=False)
print(results_df)

# Save results to a CSV file
results_df.to_csv("C:/Users/ASUS/OneDrive/DataScience/FINAL-PROJECT/data-science-Optimal-EV-station-placement-main/main_raw/data/results1/all_cities_results.csv", index=False)
```

```python
# Function to run experiment
def run_experiment(X_train, X_test, y_train, y_test):
    classifiers = all_estimators(type_filter='classifier')
    results = []
    models = {}
    for name, ClassifierClass in tqdm(classifiers):
        try:
            model = ClassifierClass()
            model.fit(X_train, y_train)
            models[name] = model
            y_pred = model.predict(X_test)

            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='macro')
            recall = recall_score(y_test, y_pred, average='macro')
            f1 = f1_score(y_test, y_pred, average='macro')
            auc = roc_auc_score(y_test, y_pred)
            balanced_accuracy = balanced_accuracy_score(y_test, y_pred)

            results.append([name, accuracy, precision, recall, f1, auc, balanced_accuracy])
        except Exception as e:
            print(f"Error occurred for {name}: {str(e)}")
    results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-score", "AUC", "Balanced Accuracy"])
    results_df = results_df.sort_values(by=['F1-score', 'AUC'], ascending=False)
    return results_df, models

# Run experiment with scaled features
result_df, models = run_experiment(X_train, X_test, y_train, y_test)

# Save individual experiment results to CSV
result_df.to_csv("C:/Users/ASUS/OneDrive/DataScience/FINAL-PROJECT/data-science-Optimal-EV-station-placement-main/main_raw/data/results/scaled_results.csv", index=False)

# Get top 5 models based on F1-score and AUC
sorted_indices = sorted(range(len(result_df['F1-score'])), key=lambda k: result_df['F1-score'][k] + result_df['AUC'][k], reverse=True)
top_5_models = [result_df['Model'][i] for i in sorted_indices[:5]]
```

```python
# Initialize variables to store top models and their accuracies
top_models = []
top_accuracies = []

# Iterate through each model
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy and classification report
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    
    # Print accuracy and classification report
    print(f"{model_name} Accuracy: {accuracy}")
    print(report)
    
    # Store accuracy and model information
    top_models.append((model_name, model))
    top_accuracies.append(accuracy)

# Find the top 5 models based on accuracy
sorted_indices = sorted(range(len(top_accuracies)), key=lambda k: top_accuracies[k], reverse=True)
top_5_models = [top_models[i] for i in sorted_indices[:5]]

# Save the predictions for the top 5 models to CSV files
for model_name, model in top_5_models:
    y_pred_top = model.predict(X_test)
    predictions_df = pd.DataFrame({
        'geometry': data['geometry'],
        'city': data['city'],
        'EV_stations': data['EV_stations'],
        f'predicted_EV_stations_{model_name}': y_pred_top
    })

    # Print lengths for troubleshooting
    print(len(data['geometry']))
    print(len(data['city']))
    print(len(data['EV_stations']))
    print(len(y_pred_top))
    
    predictions_df.to_csv(f"C:/Users/ASUS/OneDrive/DataScience/FINAL-PROJECT/data-science-Optimal-EV-station-placement-main/main_raw/data/results1/{model_name}_predicted_ev_charging_stations.csv", index=False)

# Save accuracy metrics for all models to a CSV file
accuracy_df = pd.DataFrame({
    'Model': list(models.keys()),
    'Accuracy': top_accuracies
})
accuracy_df.to_csv("C:/Users/ASUS/OneDrive/DataScience/FINAL-PROJECT/data-science-Optimal-EV-station-placement-main/main_raw/data/results1/model_accuracy_metrics.csv", index=False)
```
